# 01 — MDP Formulation
**Week 3 | RL Fundamentals**

A **Markov Decision Process** (MDP) is defined by the 5-tuple:

$$M = (\mathcal{S},\; \mathcal{A},\; \mathcal{R},\; \mathcal{T},\; \gamma)$$

| Symbol | Meaning |
|---|---|
| S | State space |
| A | Action space |
| R(s,a,s') | Reward function |
| T(s,a,s') | Transition function P(s'|s,a) |
| γ ∈ [0,1) | Discount factor |

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
np.random.seed(0)
```

## 1. Constructing a Tiny MDP by Hand
A recycling robot (from Sutton & Barto, Example 3.3):  
- **States**: {high battery, low battery}  
- **Actions**: {search, wait, recharge}  
- **Goal**: maximise cans collected

```python
# States
S = ['high', 'low']
s_idx = {s: i for i, s in enumerate(S)}

# Actions (not all available in every state but we'll handle that via masks)
A = ['search', 'wait', 'recharge']
a_idx = {a: i for i, a in enumerate(A)}

# Transition tensor: T[s, a, s'] = P(s'|s,a)
# and Reward matrix:  R[s, a] = expected reward
alpha = 0.9   # P(stay high | high, search)
beta  = 0.4   # P(stay low  | low,  search)
r_search_high = 4.0
r_search_low  = 2.0
r_wait        = 1.0

T = np.zeros((2, 3, 2))   # (states, actions, next_states)
R = np.full((2, 3), -np.inf)  # -inf for illegal actions

# From HIGH
T[0, 0, 0] = alpha;    T[0, 0, 1] = 1 - alpha;  R[0, 0] = r_search_high  # search
T[0, 1, 0] = 1.0;                                R[0, 1] = r_wait         # wait
# recharge not available from high

# From LOW
T[1, 0, 0] = 1-beta;   T[1, 0, 1] = beta;        R[1, 0] = r_search_low   # search
T[1, 1, 1] = 1.0;                                 R[1, 1] = r_wait         # wait
T[1, 2, 0] = 1.0;                                 R[1, 2] = 0.0            # recharge

print("Transition tensor shape:", T.shape, "  (states × actions × next_states)")
print("\nReward matrix R[s,a]:")
for i, s in enumerate(S):
    for j, a in enumerate(A):
        r = R[i, j]
        print(f"  R[{s},{a}] = {r if r != -np.inf else 'N/A'}")
```

## 2. Simulate One Episode Under a Random Policy

```python
def valid_actions(s):
    if s == 0:  return [0, 1]        # high: search, wait
    else:       return [0, 1, 2]     # low:  search, wait, recharge

def step(s, a):
    next_s = np.random.choice(2, p=T[s, a])
    r = R[s, a]
    return next_s, r

def simulate_episode(start=0, n_steps=20, gamma=0.95):
    s = start
    trajectory = []
    for t in range(n_steps):
        a = np.random.choice(valid_actions(s))
        next_s, r = step(s, a)
        trajectory.append({'t': t, 'state': S[s], 'action': A[a], 'reward': r, 'next': S[next_s]})
        s = next_s
    return trajectory

traj = simulate_episode()
print(f"{'t':>3}  {'State':<8} {'Action':<10} {'Reward':>7}  {'Next':>8}")
print("-" * 45)
for step_ in traj[:10]:
    print(f"{step_['t']:>3}  {step_['state']:<8} {step_['action']:<10} {step_['reward']:>7.1f}  {step_['next']:>8}")
```

Solution 1-

State ($S$): The complete configuration of the $8 \times 8$ chessboard (positions of all remaining pieces), whose turn it is, castling rights (kingside/queenside for both), en passant eligibility, and the 50-move draw counter.


Action ($A$): The set of all legal moves a player can make given the current board state according to standard chess rules.


Reward ($R$): $0$ for all intermediate moves. $+1$ for a win (checkmate or resignation), $-1$ for a loss, and $0$ for a draw (stalemate, threefold repetition, etc.). Alternatively, a dense reward function could include material differential (e.g., $+1$ for taking a pawn, $+9$ for a queen), but a sparse reward on game outcome prevents sub-optimal greedy piece-capturing behavior.

Solution 2-

If the warehouse has a maximum capacity $C$, the number of items ordered cannot cause the inventory to exceed $C$. If the current stock at morning is $s_t$, the maximum amount you can order is restricted. This alters the action space such that it becomes state-dependent:$$A(s) = \{a \in \mathbb{Z}_{\ge 0} \mid a \le C - s\}$$Any order action $a > C - s$ is rendered illegal or heavily penalized.

Solution 3-

 You can turn Chess into a POMDP by introducing Fog of War rules (often called Dark Chess). In this setup, a player can only see squares containing their own pieces and the squares their pieces can legally move to or attack. All other squares on the board remain hidden (shrouded in darkness). The state is no longer fully observable; instead, the agent receives an observation $O$ representing its visible subgrid and must maintain a probability belief distribution over where the opponent's pieces actually reside.
