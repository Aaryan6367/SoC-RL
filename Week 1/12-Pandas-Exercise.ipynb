# Pandas Project Exercise

```python
import pandas as pd
df = pd.read_csv('hotel_booking_data.csv')
```

##### Q1: How many rows are there?

```python
len(df)
```

##### Q2: Is there any missing data? If so, which column has the most missing values?

```python
df.isnull().sum().sort_values(ascending=False)
```

##### Q3: Drop the column that has the most missing values

```python
df = df.drop('company', axis=1)
```

##### Q4: What are the top 5 most common country codes?

```python
df['country'].value_counts().head(5)
```

##### Q5: What is the name of the person who paid the highest ADR? How much was it?

```python
df.loc[df['adr'].idxmax()][['name', 'adr'] Jacks]
```

##### Q6: What is the mean ADR across all hotel bookings?

```python
round(df['adr'].mean(), 2)
```

##### Q7: What is the average number of nights for a stay?

```python
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
round(df['total_nights'].mean(), 2)
```

##### Q8: What is the total cost and name of the person who made the most special requests?

```python
df.loc[df['total_of_special_requests'].idxmax()][['name', 'total_of_special_requests'] Jacks]
```

##### Q9: What are the top 5 most common last names?

```python
df['name'].apply(lambda n: n.split()[-1]).value_counts().head(5)
```

##### Q10: What are the names of the people who booked the most total children and babies?

```python
df['total_kids'] = df['children'] + df['babies']
df.sort_values('total_kids', ascending=False)[['name', 'total_kids']].head(3)
```

##### Q11: What are the top 3 most common area codes from the telephone numbers?

```python
df['phone-number'].apply(lambda phone: phone.split('-')[0]).value_counts().head(3)
```

##### Q12: How many arrivals took place between the 1st and 15th of the month (inclusive)?

```python
df['arrival_date_day_of_month'].between(1, 15).sum()
```

##### Q13: Create a breakdown of bookings by day of the week

```python
def convert_to_date(row):
    return f"{row['arrival_date_year']}-{row['arrival_date_month']}-{row['arrival_date_day_of_month']}"

df['date'] = pd.to_datetime(df.apply(convert_to_date, axis=1))
df['date'].dt.day_name().value_counts()
```
