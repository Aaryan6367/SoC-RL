### Task One: Creating data from an equation
Create arrays `m` and `E` modeling $E=mc^2$.

```python
import numpy as np
import matplotlib.pyplot as plt

m = np.linspace(0, 10, 11)
c = 3 * 10**8
E = m * (c**2)
print(m)
print(E)
```

### Part Two: Plotting E=mc^2

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(m, E, color='red', lw=5)
ax.set_title('E=mc^2')
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in Joules')
ax.set_xlim(0,10)
plt.show()
```

### Part Three (BONUS): Logarithmic representation with grid lines

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(m, E, color='red', lw=5)
ax.set_title('E=mc^2')
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in Joules')
ax.set_xlim(0,10)
ax.set_yscale('log')
ax.grid(True, which='both', axis='y')
plt.show()
```

### Task Two: Recreating curves from data points

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

##### Plot both curves on the same figure with a legend

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend()
plt.show()
```

##### Place the legend outside the bounding plot matrix

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### Recreate stacked subplots mapping

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 6))
axes[0].plot(labels, july16_2007)
axes[0].set_title("July 16 2007")
axes[1].plot(labels, july16_2020)
axes[1].set_title("July 16 2020")
plt.tight_layout()
plt.show()
```

##### BONUS CHALLENGE: Twin axis layout implementation

```python
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(labels, july16_2007, color="blue")
ax1.set_ylabel("July 16 2007", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, color="red")
ax2.set_ylabel("July 16 2020", color="red")
ax2.tick_params(axis='y', labelcolor="red")
plt.show()
```
