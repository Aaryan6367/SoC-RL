## Numpy Indexing and Selection
Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
mat = np.arange(1,26).reshape(5,5)
mat
```

##### Replicate the output block containing [[12, 13, 14, 15], [17, 18, 19, 20], [22, 23, 24, 25]]

```python
mat[2:, 1:]
```

##### Grab value 20

```python
mat[3, 4]
```

##### Grab vertical block column layout corresponding to values 2, 7, 12

```python
mat[:3, 1:2]
```

##### Grab row 4 array

```python
mat[4, :]
```

##### Grab rows index 3 and 4 blocks matrix

```python
mat[3:, :]
```

### Numpy Operations

##### Get the sum of all the values in mat

```python
mat.sum()
```

##### Get the standard deviation of the values in mat

```python
mat.std()
```

##### Get the sum of all the columns in mat

```python
mat.sum(axis=0)
```
